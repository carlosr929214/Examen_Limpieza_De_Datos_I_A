```python
############################################## Variables Categóricas #######################################################
```

```python
cols = ['CLIENTNUM','Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status',
       'Income_Category', 'Card_Category'] 
```

```python
cols
```

```python
cols_validas = []
cols_many_levels = []
cols_one_level = []

size_df = bank.shape[0]
umbral = .95

for c_ in cols:
    
    niveles = bank.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if ( niveles / size_df ) >= umbral:
        cols_many_levels.append(c_)
    elif niveles == 1:
        cols_one_level.append(c_)
    else:
        cols_validas.append(c_)
```

```python
############################################## Columna Education_Level #######################################################
```

```python
bank.loc[:, "Education_Level"]
```

```python
levels1 = bank.groupby(["Education_Level"], as_index = False ).agg({"CLIENTNUM":{"count"}})
levels1.columns = levels1.columns.droplevel(1)
levels1
```

```python
levels1['Percentage'] = levels1['CLIENTNUM'] / bank.shape[0]
levels1
```

```python
levels1.sort_values( ['Percentage'], ascending=False )
```

```python
levels1['Education_Level_t'] = 'others'
levels1
```

```python
levels1['Education_Level_t'] = levels1.apply( lambda r: r['Education_Level_t'] if r['Education_Level'] == "Unknown" else r['Education_Level'] , axis = 1 )
levels1
```

```python
levels1 = levels1.loc[ : , ["Education_Level","Education_Level_t"] ]
levels1
```

```python
Banco1 = bank.merge( levels1, on =["Education_Level"], how="left")
Banco1
```
